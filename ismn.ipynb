```python
import pandas as pd
import os
import glob
from tqdm import tqdm

# Define the root directory containing all the subdirectories
root_dir = 'TxSON/'

# Find all .stm files in the subdirectories of root_dir
file_paths = glob.glob(os.path.join(root_dir, '*', '*.stm'))

# Initialize a list to collect DataFrames (if needed)
dataframes = []

# Process each file
for file_path in tqdm(file_paths):
    # Read the metadata line separately (assuming it's the first line of the file)
    with open(file_path, 'r') as file:
        metadata_line = file.readline()
        lat_lon = metadata_line.split()[3:5]  # Adjust indices based on actual structure
        latitude = float(lat_lon[0])
        longitude = float(lat_lon[1])

    # Read the remaining time series data
    df = pd.read_csv(
        file_path,
        delim_whitespace=True,
        skiprows=1,  # Skip the metadata line
        names=['Date', 'Time', 'Value', 'Unit1', 'Unit2'],
        parse_dates={'Date_Time': [0, 1]}
    )

    # Add latitude and longitude as new columns
    df['Latitude'] = latitude
    df['Longitude'] = longitude

    # Optional: Append to the list of DataFrames
    dataframes.append(df)

    # Optional: Display each DataFrame (or you can handle it differently)
    print(df)

# At this point, 'dataframes' contains all the data from all processed files
# You can now perform further analysis or save them as needed
```

```python
# Combine all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd

# # Create a sample DataFrame
# data = {
#     'Latitude': [30.3423, 30.2836, 30.4000, 30.3500],
#     'Longitude': [-98.8679, -98.7417, -98.7000, -98.6500]
# }
# df = pd.DataFrame(data)

# Getting unique latitude and longitude
unique_coords = df.drop_duplicates(subset=['Latitude', 'Longitude'])

# Set up the plot with Cartopy
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-99, -98.5, 30.2, 30.5], crs=ccrs.PlateCarree())

# Add geographic features
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)

# Plot the unique coordinates
ax.scatter(unique_coords['Longitude'], unique_coords['Latitude'],
           color='red', s=10, label='Locations', transform=ccrs.PlateCarree())

# Add gridlines and labels
gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
gl.top_labels = False
gl.right_labels = False

# Show the plot
plt.legend()
plt.show()
```
